```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
#your code here
print(spaceship.isnull().sum())
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
# Drop rows with missing values
spaceship = spaceship.dropna()

# Verify again
print(spaceship.isnull().sum())
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
display(spaceship)
display(spaceship['Cabin'])
```

```python
# Transform 'Cabin' to extract the first letter representing the deck
spaceship['Cabin'] = spaceship['Cabin'].str[0]

# Check
print(spaceship['Cabin'].unique())
```

```python
display(spaceship)
```

- Drop PassengerId and Name

```python
#your code here
spaceship = spaceship.drop(['Name'], axis=1)
spaceship = spaceship.drop(['PassengerId'], axis=1)
```

- For non-numerical columns, do dummies.

```python
#your code here
# Convert categorical columns to dummy variables
spaceship = pd.get_dummies(spaceship, columns=['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP'], drop_first=True)

# Check the resulting dataframe
print(spaceship.head())
```

**Perform Train Test Split**

```python
#your code here
```

```python
# Define X (features) and y (target)
X = spaceship.drop('Transported', axis=1)
y = spaceship['Transported'].astype(int)

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize KNN
knn = KNeighborsClassifier()

# Train the model
knn.fit(X_train, y_train)
```

- Evaluate your model's performance. Comment it

```python
#your code here
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Print classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

Key Metrics:

- Accuracy is 77%. This indicates that the model correctly predicted the outcome for 77% of the cases in the test dataset. While not perfect, it’s a good starting point.

Confusion Matrix:
- True Positives (501): The model correctly predicted 501 instances as “Transported”.
- True Negatives (516): The model correctly predicted 516 instances as “Not Transported”.
- False Positives (145): The model predicted 145 instances as “Transported”, but they were actually “Not Transported”.
- False Negatives (160): The model predicted 160 instances as “Not Transported”, but they were actually “Transported”.
